# Question 4.	Find your mean squared error and R2 values. Be sure to indicate what these tell you. You can import these from sklearn.metrics . 

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error

r2_score(y_test, y_pred)
```

```python
mean_squared_error(y_test, y_pred)
```

The r2 score gives you the same result as the accuracy score from above.  R2 is the proportion of variance in the dependent variable that is predicted by the independent variable.  In this case, the dependent variable in the water temperature and the independent variable is the salinity. So, these results show that the salinity of the water can not reliably be used to predict the water temperature.  Linear regression predicts a line with the formula y=mx+b that will best fit the data points.  How far each data point is from that line is called the residual. Mean-Squared error is the mean of the squares of the residuals. They are squared since some are negative and some are positive. This gives an idea of how far away the prediction is from the test values. In our case, the number is high, meaning the prediction is not very good.
