# Question 1: Use Splinter and/or Beautiful Soup to scrape https://www.dndbeyond.com/classes, and pull out all class names. Store these in a dataframe.

```python
#import necessary modules
from splinter import Browser
from webdriver_manager.chrome import ChromeDriverManager
import bs4
import pandas as pd
```

```python
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
#set an empty list all_classes
all_classes=[]
#set the url we want to scrape
url = 'https://www.dndbeyond.com/classes'
#visit the selected url
browser.visit(url)
#define an empty dictionary
class_dict={}
#use beautiful soup for parse website
soup=bs4.BeautifulSoup(browser.html, 'html.parser')
#find all class names under the class 'listing-card__header-primary'
all_divs = soup.find_all('div',class_='listing-card__header-primary')
#run through each div under the listing card header primary
for div in all_divs:
    try:
        #get the class name for each class
        classes= div.find('h3',class_='listing-card__title').text
        #remove the blank spaces before and after the class name
        classes=classes.strip()
        #create a dictionary of the classes
        class_dict = {'class':classes}
            
        #append the dictionary to the list
        all_classes.append(class_dict)
            
    except AttributeError as e:
        print(e)
    
#create a dictionary from the list of classes
df = pd.DataFrame(all_classes)
df.head(-1)
```

# Question 2: Now, alter 1 to click on each class name to go to the more detailed page, and pull out the first paragraph about each class. Store these in a dataframe. Your dataframe should have two columns.

```python
#create an empty list all_classes
all_classes=[]
#set the url to the url we want to scrape
url = 'https://www.dndbeyond.com/classes'
#open a browser for that url
browser.visit(url)
#create an empty dictionary
class_dict={}
#use beautiful soup to parse website
soup=bs4.BeautifulSoup(browser.html, 'html.parser')
#use beautiful soup to find all class names under the class 'listing-card__header-primary'
all_divs = soup.find_all('div',class_='listing-card__header-primary')
for div in all_divs:
    
    #find the class name under the class listing card title
    classes= div.find('h3',class_='listing-card__title').text
    #remove extra spaces before and after class name
    classes= classes.strip()

    #run through each class name, opening up the url for that specific class.  Note: Artificer and Blood Hunter are a different category and should be excluded. 
    if classes != 'Artificer' and classes != 'Blood Hunter':
        url = f'https://www.dndbeyond.com/classes/{classes}'
        #assign a new browser specific to that url
        browser.visit(url)
        #use beautiful soup to parse that specific url
        soup=bs4.BeautifulSoup(browser.html, 'html.parser')
        #find the first paragraph about that specific class found under the class core styles core body
        description = soup.find('p', class_='Core-Styles_Core-Body').text
        #fill in a dictionary with the class name and description
        class_dict = {'Class':classes, 'Description':description}
        
        #append the dictionary to the list all_classes
        all_classes.append(class_dict)

    
#make a dataframe from the list
df = pd.DataFrame(all_classes)
df.head(-1)
```
