# Linear vs. Rank Correlation

Linear correlation measures the proximity of the mathematical relationship between variables or dataset features to a linear function.  If the relationship between the two variables is very close to a linear function than the correlation will be high. 

Rank correlation compares the ranks or the orderings fo the data related to two variables or dataset features. If the orderings are similar, then the correlation is strong.  Rank correlation is only concerned with the order of values, not with the particular values. Example, as x increases, does y increase?  The relationship does not have to be linear. Spearman correlation is an example.  You only care about the direction of the relationship.  


```python
# Create a scatter plot showing the relationship between calculated_host_listings_count and availability_365 in the belize AirBnb data
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

belize_listings_path = "../week_05/belize_airbnb_listings.csv"
belize_df = pd.read_csv(belize_listings_path)
belize_df.head()

plt.scatter(belize_df['calculated_host_listings_count'],belize_df['availability_365'])
plt.title("Belize AirBnb Listings Count vs. Availability")
plt.xlabel("Listings Count")
plt.ylabel("Availability 365")
```

```python
np.corrcoef(belize_df['calculated_host_listings_count'],belize_df['availability_365'])
```

The correlation coefficient between these two variables is very small.  This shows that there is virtually no correlation between the number of listings that a host has for an AirBnb (which is usually quite low) and the number of days that the listings are available.
